## A fancier plot with real data

Python is not limited to toy examples. It can also download real data
from a public open-data portal and turn it into a plot.

In the next cell, we download a dataset from NYC Open Data and plot:
- total New York City water consumption
- per-capita water use

Run the cell and inspect the graph.


```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://data.cityofnewyork.us/api/views/ia2d-e54m/rows.csv?accessType=DOWNLOAD"
df = pd.read_csv(url)

fig, ax1 = plt.subplots(figsize=(10, 5))
ax1.plot(df["Year"], df["NYC Consumption(Million gallons per day)"], linewidth=2)
ax1.set_xlabel("Year")
ax1.set_ylabel("NYC water consumption (million gallons/day)")
ax1.set_title("New York City Water Consumption Over Time")

ax2 = ax1.twinx()
ax2.plot(df["Year"], df["Per Capita(Gallons per person per day)"], linestyle="--", linewidth=2)
ax2.set_ylabel("Per capita use (gallons/day)")

fig.tight_layout()
plt.show()
```

Try this

Change the code and run it again.

Ideas:
- plot only one of the two lines
- change the title
- print `df.head()` to inspect the first rows of data
- change `figsize=(10, 5)` to another size

10b. Working with Gemini


Colab can also be used with Gemini, Google's AI system.

For this course, use Gemini as a helper, not as a substitute for your own work.

Good uses:
- ask for an explanation of an error message
- ask what a short piece of code does
- ask for a small variation on code you already understand
- ask for hints instead of full solutions

Poor uses:
- copying answers without running them
- submitting code you cannot explain
- asking for a full assignment solution and pasting it unchanged


Prompt ideas

If you try Gemini, use prompts like these:

- "Explain this error message in simple language."
- "What does this loop do, line by line?"
- "Change this code so it prints numbers from 1 to 10."
- "Give me a hint, but do not give me the full solution."

Always test the response by running the code yourself.
