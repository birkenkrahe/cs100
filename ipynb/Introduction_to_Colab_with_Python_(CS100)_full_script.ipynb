## Python Can Do Arithmetic

Computers are fancy calculators. Run the cells and check the results:

````python
print(2 + 3)
print(10 - 4)
print(6 * 7)
print(8 / 2)
````

> The argument of `print` here is not a string but an **expression** — something the computer evaluates using numbers. All results except the last are **integers**; division always returns a **floating-point** number.

There are more common binary operators:

````python
print(2 ** 5)   # exponentiation
print(17 % 3)   # remainder (modulo)
print(17 // 3)  # integer (floor) division
````

> Floor division `a//b` returns `floor(a/b)`. For negative numbers, floor rounds toward negative infinity: `-17 // 3 = -6` (not -5). Python preserves the invariant: `(a//b)*b + (a%b) == a`.
>
> `#` is the **comment** sign — everything after it is ignored.

Before running the next cell, predict the result:

````python
print(3 + 4 * 5)  # prints 23 = 3 + 20
````

````python
print((3 + 4) * 5)  # prints 7 * 5 = 35
````

> Evaluating expressions follows the PEMDAS rule: **P**arentheses > **E**xponentiation > **M**ultiplication > **D**ivision > **A**ddition > **S**ubtraction.

## Words and Numbers Are Different

Python distinguishes between text and numbers at the user level. At the machine level, only numbers exist — the letter `A` corresponds to ASCII code `65`:

````python
print(ord('A'))        # converts character to ASCII number
print(chr(65))         # converts ASCII number to character
print(type(ord('A')))  # it's an int
print(type(chr(65)))   # it's a str
````

````python
print(33)
print("33")
````

No difference in display — but the `type` is different:

````python
print(type(33))    # integer
print(type("33"))  # string
````

Introducing operations changes the picture:

````python
print(2 + 3)
print("2" + "3")
````

> `2 + 3` is arithmetic; `"2" + "3"` uses the same symbol for **string concatenation**.

## A Small "Fancy" Example

Python can make simple graphics.

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]   # list
y = [1, 4, 9, 16, 25]  # another list

plt.plot(x, y)  # plot y = f(x)
plt.show()
````

> - The first line imports a **library** so that we can plot graphics.
> - `x` and `y` are **lists**; their values form `(x,y)` pairs: `(1,1)`, `(2,4)`, `(3,9)` etc.
> - `plot()` draws a line between these pairs.
> - `show()` displays the plot in the notebook.

## A Fancier Plot with Real Data

Here we download real data from a public open data portal and plot total New York City water consumption and per-capita water use.

````python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://data.cityofnewyork.us/api/views/ia2d-e54m/rows.csv?accessType=DOWNLOAD"
df = pd.read_csv(url)

fig, ax1 = plt.subplots(figsize=(10, 5))
ax1.plot(df["Year"],
         df["NYC Consumption(Million gallons per day)"],
         linewidth=2)
ax1.set_xlabel("Year")
ax1.set_ylabel("NYC water consumption [mio gallons/day]")
ax1.set_title("New York City Water Consumption 1960-2024")

ax2 = ax1.twinx()
ax2.plot(df["Year"],
         df["Per Capita(Gallons per person per day)"],
         linestyle="--",
         linewidth=2)
ax2.set_ylabel("Per capita water use [gallons/day]")
fig.tight_layout()
plt.show()
````

> **Interpretation:**
> - Consumption rises from 1960 to ~1980 (1,500 mio gal/day)
> - Sustained decline from 1980 to 1,000 mio gal/day
> - NYC population grew from 7.78M (1960) to 8.48M (2024)
> - Dip in 1965: drought (external shock)
> - Both curves track together: population growth is not the driver
> - Post-2000 decline is structural: metering, appliance efficiency, de-industrialization

## Repetition with a Loop

Data usually comes in bulk. To process many data points in sequence, programming languages "loop" over the data.

In the following `for` loop, a variable `i` ranges over five values generated by the `range()` function:

````python
for i in range(5):
    print("Python is running", i)
````

Like C, Python starts counting at 0, and the `range()` argument is an upper bound (not included).

## Random Numbers

Python's `random` module uses the Mersenne Twister, a pseudo-random number generator (PRNG) — deterministic given a seed. Its numbers can be reproduced exactly for any given seed. There are Python libraries that harvest external entropy sources, getting closer to true random number generation.

This code uses a `for` loop to generate 5 integer random numbers between 1 and 10:

````python
import random

for i in range(5):
    print(random.randint(1, 10))
````
