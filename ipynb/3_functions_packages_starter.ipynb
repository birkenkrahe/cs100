## Step 1: Anatomy of a function

A function groups reusable code under a name. Here is the smallest useful example — a greeting:

Every function definition has the same parts:

| Part | In this example | Comment |
|------|----------------|---------|
| keyword | `def` | Protected |
| name | `greeting` | Namespace |
| parameters | `greet`, `name` | No default |
| colon | `:` | Delimiter |
| body | `print(greet + ", " + name)` | Scope |
| docstring | the triple-quoted string | Comment |
| call | `greeting("Hello", "Marcus")` | Variables |
| arguments | `"Hello"`, `"Marcus"` | Strings |

- **Parameters** belong to the **definition**; **arguments** belong to the **call**.
- The **body** is indented — Python uses indentation to know what's inside.
- The **docstring** is optional but good practice: `help(greeting)` shows it.

## Step 2: Traceback and pythontutor.com

When your code crashes, Python shows you the whole chain of calls that led there, in the order of calls (stacked). This is called a **traceback**.

Copy the `greeting` function from Step 1. Change `name` to `names` in the body — a one-letter typo — and run it:

````python
def greeting(greet, name):
    print(greet + ", " + names + "!")  # NameError: names not defined

def main():
    greeting("Hello", "Marcus")

if __name__ == "__main__":
    main()
````

Python prints a **traceback** — read it bottom to top:

```
NameError: name 'names' is not defined    ← what went wrong
  greeting(...)                            ← where it happened
  main()                                   ← who called it
  <module>                                 ← the top-level script
```

- The error is always at the bottom — start there.
- Each line above it is one step back up the call chain (the "stack").
- Python often suggests a fix: *Did you mean: 'name'?*

Open [pythontutor.com](https://pythontutor.com/python-debugger.html), paste the working `greeting` code, and step through it. You will see `__main__` → `main()` → `greeting()` appear as live stack frames — the traceback made visual.

## Bonus: What is `__main__`?

The textbook mentions it at the beginning of chapter 3:

> "`__main__` is the name of the module that contains the function `print_lyrics`." — But what does that mean?

> Every Python file gets an automatic label called `__name__`. The double underscores mean that it's Python stuff, not user stuff.
>
> When you run code directly in Colab, or in a terminal, Python sets `__name__` to the special value `__main__`, and all functions you define live in it.

## Summary

- `def` defines a function; the body must be indented
- Parameters are placeholders in the definition; arguments are the values passed at the call
- A docstring documents what the function does; `help()` displays it
- A traceback lists the call chain from `<module>` down to the error; read it bottom to top
- `__main__` is the name Python gives to the currently running module
- `if __name__ == "__main__":` guards code that runs only when the file is executed directly, not when imported
- `import numpy as np` loads NumPy; `np.array()` converts a list to an array
- Array arithmetic applies to every element at once — no loop needed
- `np.min()`, `np.max()`, `np.mean()` compute statistics on the whole array

## Glossary

| Term | Meaning |
|------|---------|
| `def` | Keyword that starts a function definition |
| parameter | Variable name in the function definition |
| argument | Value passed to the function at the call site |
| docstring | String literal at the top of a function body; shown by `help()` |
| module | A Python file that can be imported into another script |
| `__main__` | The value of `__name__` when a file is run directly |
| `if __name__ == "__main__":` | Guard that runs code only when executed directly, not when imported |
| traceback | The call chain Python prints when a runtime error occurs |
| stack frame | One function's local variables and position in the chain |
| `import` | Makes a package available in the current script |
| `numpy.ndarray` | NumPy's array type — stores elements of one type |
| vectorized | Operation that applies to every element of an array at once |
